```python
import pandas

df = pandas.read_csv("estudiantes_notas.csv")
```

```python
print(df.head())
print(df.tail())
print("Columnas:", len(df.columns))
print("Filas", len(df))
print(df.isnull().sum())

df_minmax = (df - df.min()) / (df.max() - df.min())
print("Normalización Min-Max:\n", df_minmax)

# Normalización Z-Score
df_zscore = (df - df.mean()) / df.std()
print("Normalización Z-Score:\n", df_zscore)
```

```python
nota_sobre_90 = df[df['Nota_Matemáticas'] > 90] 
print(nota_sobre_90)
# Promedio mayor 85 por notas
df["Promedio_Notas"] = df[['Nota_Matemáticas', 'Nota_Lenguaje', 'Nota_Ciencias']].mean(axis=1)
promedio_notas = df[df["Promedio_Notas"] > 85]
print(promedio_notas)
```

```python
# ordenado por Promedio_Notas de forma descendente.
promedio_descendente = df.sort_values(by="Promedio_Notas", ascending=False)
print(promedio_descendente)

print(df[["Estudiante_ID", "Promedio_Notas"]])

menos_10 = df[df['Nota_Matemáticas'] < 10]
print(menos_10)

df.to_csv("promedio_estudiantes.csv")
promedio_notas.to_csv("mayor_85.csv")
```
